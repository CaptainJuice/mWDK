```python

psi_lst_mWDK = [32]
interclass_noise_rate_lst = [0.1/16,0.1/8,0.1/4,0.1/2,0.1]
intraclass_noise_rate_lst = [0.1,0.2,0.4]

save_folder = 'mWDK_noise_effect_results'
save_folder_2 = 'WL_noise_effect_results'
save_folder_3 = 'WDK_noise_effect_results'
dataset = 'acm'
h_ = 3
xticks_lst = ['$0.1/16$', '$0.1/8$', '$0.1/4$', '$0.1/2$', '$0.1$']

color_lst = ['tab:blue','tab:orange','tab:green','tab:red','tab:purple','tab:brown','tab:pink','tab:gray','tab:olive','tab:cyan']

for psi in psi_lst_mWDK:
    plt.figure()
    for k,intraclass_noise_rate in enumerate(intraclass_noise_rate_lst):
        plot_data = list() # mWDK
        plot_data_2 = list() # WL
        plot_data_3 = list() # WDK
        for interclass_noise_rate in interclass_noise_rate_lst:
            file_name = dataset+'_psi_'+str(psi)+'_h_'+str(h_)+'_inter_'+str(interclass_noise_rate)+'_intra_'+str(intraclass_noise_rate)+'.mat'
            file_name_2 = dataset+'_psi_'+str(-1)+'_h_'+str(4)+'_inter_'+str(interclass_noise_rate)+'_intra_'+str(intraclass_noise_rate)+'.mat'
            file_name_3 = dataset+'_psi_'+str(psi)+'_h_'+str(2)+'_inter_'+str(interclass_noise_rate)+'_intra_'+str(intraclass_noise_rate)+'.mat'
            nmi_iter = scipy.io.loadmat(save_folder+os.sep+file_name)['nmi'] # mWDK
            nmi_iter_2 = scipy.io.loadmat(save_folder_2+os.sep+file_name_2)['nmi'] #WL
            nmi_iter_3 = scipy.io.loadmat(save_folder_3+os.sep+file_name_3)['nmi'] #WDK
            nmi = nmi_iter[0,-1]
            nmi_2 = nmi_iter_2[0,-1]
            nmi_3 = nmi_iter_3[0,-1]
            plot_data.append(nmi)
            plot_data_2.append(nmi_2)
            plot_data_3.append(nmi_3)

        # plt.plot(list(range(len(interclass_noise_rate_lst))),plot_data, label='(mWDK) intraclass noise rate='+str(intraclass_noise_rate),
        #          marker='*', markersize=7.5,linewidth=1,color=color_lst[k])
        plt.plot(list(range(len(interclass_noise_rate_lst))),plot_data_2, label='(WL) intraclass noise rate='+str(intraclass_noise_rate),
                 marker='*', markersize=7.5,linewidth=1,color=color_lst[k])
        # plt.plot(list(range(len(interclass_noise_rate_lst))),plot_data_3, label='(WDK) intraclass noise rate='+str(intraclass_noise_rate),
        #          marker='*', markersize=7.5,linewidth=1,color=color_lst[k])

    leng = ['intraclass noise rate = '+str(x) for x in intraclass_noise_rate_lst]
    plt.legend(leng,fontsize=14)
    plt.xlabel('interclass noise rate',fontsize=14)
    plt.ylabel('NMI',fontsize=14)
    plt.xticks(list(range(len(interclass_noise_rate_lst))),xticks_lst,fontsize=12)
    plt.yticks(fontsize=12)
    plt.ylim(0,1)
    plt.show()
```
